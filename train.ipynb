```python

folder0 = 'D:/LearnByself/Python/Age/18'
folder1 = 'D:/LearnByself/Python/Age/19'
folder2 = 'D:/LearnByself/Python/Age/20'
folder3 = 'D:/LearnByself/Python/Age/21'
folder4 = 'D:/LearnByself/Python/Age/22'
folder5 = 'D:/LearnByself/Python/Age/23'
folder6 = 'D:/LearnByself/Python/Age/24'
folder7 = 'D:/LearnByself/Python/Age/25'
folder8 = 'D:/LearnByself/Python/Age/26'
folder9 = 'D:/LearnByself/Python/Age/27'
folder10 = 'D:/LearnByself/Python/Age/28'
folder11 = 'D:/LearnByself/Python/Age/29'
folder12 = 'D:/LearnByself/Python/Age/30'
folder13 = 'D:/LearnByself/Python/Age/31'
folder14 = 'D:/LearnByself/Python/Age/32'
folder15 = 'D:/LearnByself/Python/Age/33'
folder16 = 'D:/LearnByself/Python/Age/34'
folder17 = 'D:/LearnByself/Python/Age/35'
folder18 = 'D:/LearnByself/Python/Age/36'
folder19 = 'D:/LearnByself/Python/Age/37'
folder20 = 'D:/LearnByself/Python/Age/38'
folder21 = 'D:/LearnByself/Python/Age/39'
folder22 = 'D:/LearnByself/Python/Age/40'
folder23 = 'D:/LearnByself/Python/Age/41'
folder24 = 'D:/LearnByself/Python/Age/42'
folder25 = 'D:/LearnByself/Python/Age/43'
folder26 = 'D:/LearnByself/Python/Age/44'
folder27 = 'D:/LearnByself/Python/Age/45'
folder28 = 'D:/LearnByself/Python/Age/46'
folder29 = 'D:/LearnByself/Python/Age/47'
folder30 = 'D:/LearnByself/Python/Age/48'
folder31 = 'D:/LearnByself/Python/Age/49'
folder32 = 'D:/LearnByself/Python/Age/50'
folders=[folder0,folder1,folder2,folder3,folder4,folder5,folder6,folder7,folder8,folder9,folder10,folder11,folder12,folder13,folder14,folder15,folder16,folder17,folder18,folder19,folder20,folder21,folder22,folder23,folder24,folder25,folder26,folder27,folder28,folder29,folder30,folder31,folder32]
photos, labels = list(), list()
for i in range (32):
 for file in listdir(folders[i]):
    output=i
    photo = load_img(folders[i]+'/'+file, target_size=(128,128))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

photos = asarray(photos)
labels = asarray(labels)
print(photos.shape, labels.shape)
save('D:/LearnByself/Python/Age/age_photos.npy',photos)
save('D:/LearnByself/Python/Age/age_labels.npy',labels)
x_train, x_test, y_train, y_test = train_test_split(photos, labels, test_size=0.3, random_state=2)
x_train /= 255
x_test /= 255
y_train = keras.utils.np_utils.to_categorical(y_train)
y_test = keras.utils.np_utils.to_categorical(y_test)
```

```python
model = Sequential()
model.add(Conv2D(64,(5,5), activation = 'relu',
kernel_initializer='he_uniform',padding='same',input_shape=(128,128,3)))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(128,(5,5), activation = 'relu', kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(256,(5,5), activation = 'relu',
kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(512,(5,5), activation = 'relu',
kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(512, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(32, activation='softmax'))
model.summary() 
```

```python
opt = SGD(lr=0.01, momentum=0.9)
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
model.compile(optimizer=opt,loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, batch_size=64, epochs=100, verbose=1, validation_data=(x_test, y_test), callbacks=[callback])
model.save('Age.h5')
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('epoch')
plt.legend(['Train','Validation'], loc='upper left')
plt.show()
```
